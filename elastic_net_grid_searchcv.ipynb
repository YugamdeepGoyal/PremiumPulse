```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder, StandardScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
from sklearn.linear_model import LinearRegression, ElasticNet
from load_data import get_dataset
from pipeline import get_preprocessor
```

```python
X_train, X_test, y_train, y_test = get_dataset()
```

```python
preprocessor = get_preprocessor()
```

```python
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("model", ElasticNet(max_iter=300000))
])
```

```python
params_grid = {
    "model__alpha": [0.001, 0.01, 0.1, 1, 10, 100],
    "model__l1_ratio": [0.1, 0.5, 0.9, 0.95, 0.99]
}
```

```python
grid = GridSearchCV(pipe, param_grid=params_grid, cv=5, scoring="neg_mean_absolute_error", verbose=1, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
grid.best_params_
```

```python
print("Training")
print("R2 score", r2_score(y_train, grid.predict(X_train)))
print("Mean absolute error", mean_absolute_error(y_train, grid.predict(X_train)))
print("Mean squared error", mean_squared_error(y_train, grid.predict(X_train)))
print("Root mean squared error", root_mean_squared_error(y_train, grid.predict(X_train)))
```

```python
print("Testing")
print("R2 score", r2_score(y_test, y_pred))
print("Mean absolute error", mean_absolute_error(y_test, y_pred))
print("Mean squared error", mean_squared_error(y_test, y_pred))
print("Root mean squared error", root_mean_squared_error(y_test, y_pred))
```

```python
plt.title("Residual Plot of Elastic Net with Grid Search CV")
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='--')
plt.tight_layout()
plt.savefig("images/residual_plot_of_elastic_net_grid_search.png", dpi=1200)
plt.show()
```
