```python
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
from sklearn.svm import SVR
from load_data import get_dataset
from pipeline import get_preprocessor
```

```python
X_train, X_test, y_train, y_test = get_dataset()
```

```python
preprocessor = get_preprocessor()
```

```python
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("model", SVR())
])
```

```python
params_grid = [
    {
        'model__kernel': ['linear'],
        'model__C': [0.1, 1, 10, 100],
        'model__epsilon': [0.01, 0.1, 0.2, 0.5]
    },
    {
        'model__kernel': ['rbf'],
        'model__C': [0.1, 1, 10, 100, 1000],
        'model__gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
        'model__epsilon': [0.01, 0.1, 0.2, 0.5]
    },
    {
        'model__kernel': ['poly'],
        'model__degree': [2, 3, 4],
        'model__C': [0.1, 1, 10, 100],
        'model__epsilon': [0.01, 0.1, 0.2]
    }
]
```

```python
grid = RandomizedSearchCV(pipe, param_distributions=params_grid, cv=5, scoring="neg_mean_absolute_error", verbose=1, n_iter=100, n_jobs=-1, random_state=42)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
print("Training")
print("R2 Score", r2_score(y_train, grid.predict(X_train)))
print("MAE", mean_absolute_error(y_train, grid.predict(X_train)))
print("MSE", mean_squared_error(y_train, grid.predict(X_train)))
print("RMSE", root_mean_squared_error(y_train, grid.predict(X_train)))
print("\n")
print("Testing")
print("R2 Score", r2_score(y_test, y_pred))
print("MAE", mean_absolute_error(y_test, y_pred))
print("MSE", mean_squared_error(y_test, y_pred))
print("RMSE", root_mean_squared_error(y_test, y_pred))
```

```python
grid.best_params_
```

```python
plt.title("Residual Plot of SVM")
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='--')
plt.tight_layout()
plt.savefig("images/residual_plot_of_svm.png", dpi=1200)
plt.show()
```
